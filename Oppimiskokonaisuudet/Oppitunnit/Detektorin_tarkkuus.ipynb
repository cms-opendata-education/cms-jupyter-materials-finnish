```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data=pd.read_csv("http://opendata.cern.ch/record/304/files/dielectron.csv")
data.head()
len(data)
```

```python
plt.hist(data.M, bins=100) 
plt.xlabel('Invariantti massa [GeV]')
plt.ylabel('Tapahtumien lukumäärä')
plt.title('Kahden myonin invariantti massa\n')
plt.show()
```
